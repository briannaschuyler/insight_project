```python
%matplotlib inline

from types import GeneratorType
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

#import numpy as np
import googlemaps
import math
import urllib
```

```python
# Any functions we're going to need to get this done...
def Pixels2LatLng(px_row, px_col, img_center_lat, img_center_lng, img_edge_size, img_zoom):
    """Converts a pixel row and column in a static map image to latitude and longitude."""
    # Find the center pixel of the image--since we're just using square images, we can just calculate one number.
    CenterPixel = int(img_edge_size / 2)
    # Find the number of degrees of latitude or longitude spanned by each pixel.
    ZoomScale = 2.**(img_zoom - 2)
    WholeEarthLatRes = math.pi * (180. / math.pi) / img_edge_size
    ImgLatRes = WholeEarthLatRes / ZoomScale
    WholeEarthLngRes = 2. * math.pi * (180. / math.pi) / img_edge_size
    ImgLngRes = WholeEarthLngRes * math.sin(img_center_lat * math.pi / 180.) / ZoomScale
    # Calculate the number of pixels from the center.
    pxRowDist = px_row - CenterPixel
    pxColDist = px_col - CenterPixel
    # Now Scale this distance by the resolution and reutrn them both.
    pxLat = img_center_lat + (pxColDist * ImgLatRes)
    pxLng = img_center_lng + (pxRowDist * ImgLngRes)
    return [pxLat, pxLng]
```

```python
# Set the zoom level and what we want the edge size to be.
zoom = 14
StaticImageEdgeSize = 640
# Construct the map URL, and fetch it using urllib.
MapImgageURL = "https://maps.googleapis.com/maps/api/staticmap?center=%s,%s&zoom=%i&size=%ix%i&sensor=false" % (GeocodeLat, GeocodeLng, zoom, StaticImageEdgeSize, StaticImageEdgeSize)
MapImageReturn = urllib.urlretrieve(MapImgageURL, "staticmap.png")[0]
GMImageArray = mpimg.imread(MapImageReturn)
# Find the corners of the image so we can plot it nicely and with the right aspect ratio.
TopLftCorner = Pixels2LatLng(0, 0, 
                             GeocodeLat, GeocodeLng, 
                             StaticImageEdgeSize, zoom)
BotRgtCorner = Pixels2LatLng(StaticImageEdgeSize - 1, StaticImageEdgeSize - 1, 
                             GeocodeLat, GeocodeLng, 
                             StaticImageEdgeSize, zoom)
print "The Lat, Lng of the center is:", GeocodeLat, GeocodeLng
print "The top left corner is at:", TopLftCorner[0], TopLftCorner[1]
```
