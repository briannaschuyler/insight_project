```python
# Give list of column names in the latlng dataframe
list(latlng.columns.values)

#
plt.figure(figsize=(20*cos(.65),20))
plt.scatter(latlng['AddLng'],latlng['AddLat'])

# Plot Philz coffee on Middlefield and Loma Verde in red
plt.scatter(-122.1314639, 37.4295785, color='red')
# Jordan Middle School (just north of Oregon Expressway)
plt.scatter(-122.1340302, 37.4398407,color='darkorange')
# Plot Palo Alto Gold Course (east of 101 and north corner of plot)
plt.scatter(-122.1274857,37.4527879,color='green')
# Plot Insight Building
plt.scatter(-122.1432647,37.4263312,color='yellow')
# Plot 994 Loma Verde (home)
plt.scatter(-122.1158637,37.4346362,color='pink')

print('Min Lat: '+str(min(latlng['AddLat'])))
print('Min Lng: '+str(min(latlng['AddLng'])))
print('Max Lat: '+str(max(latlng['AddLat'])))
print('Max Lng: '+str(max(latlng['AddLng'])))


# Philz 37.4295785,-122.1314639
# Goodwill o 37.4110713,-122.1353263
```
